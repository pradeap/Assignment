```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm', 

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
df['FlightNumber'] = (df['FlightNumber'].fillna(0) +10).astype(int)
df
```

```python
df[['From','To']] = df.From_To.str.split("_",expand=True)
df
```

```python
df= df.drop("From_To",axis=1)  # delete the From_To column
df
```

```python
df.columns
```

```python
df = df[['From', 'To','FlightNumber', 'RecentDelays', 'Airline']]  # rearranging the column
df
```

```python
df['From'] = df['From'].str.capitalize()
df['To'] = df['To'].str.capitalize()
df
```

```python
df1 = pd.DataFrame(df.RecentDelays.values.tolist(),index= df.index)
df1
```

```python
df1 = df1.rename(columns = {0:'Delay_1',1:'Delay_2',2:'Delay_3'})
df1
```

```python
df  = pd.concat([df,df1],axis=1)
df
```

```python
df = df.rename(columns={'RecentDelays': 'Delays'})
df
```
