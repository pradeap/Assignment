```python
from sklearn.model_selection import train_test_split
from xgboost.sklearn import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,f1_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=120)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python

params = {
    'objective': 'binary:logistic',
    'max_depth': 2,
    'learning_rate': .5,
    'silent': True,    # would be boolean in sklearn
    'n_estimators': 100
}
```

```python
xgb = XGBClassifier(objective='binary:logistic',max_depth = 2,learning_rate = 0.5,n_estimators=100).fit(x_train,y_train)
```

```python
y_pred = xgb.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
f1_score(y_test,y_pred)
```
